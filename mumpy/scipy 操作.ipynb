```python
integrate.quad(lambda x:np.sin(x),0,np.pi)
# 结果的第一个元素是 数值 ，第二个元素是包含误差的上限
```

```python
from scipy.integrate import dblquad
area=dblquad(lambda x,y:x*y,0,0.5,lambda x:0,lambda x: 1-2*x)
area
```

```python
n=5
def f(t,x):
    return np.exp(-x*t)/t**n
    pass
u=integrate.nquad(f,[[1,np.inf],[0,np.inf]])
res=1/5
err=res-u[0]
print(abs(err))
```

```python
from scipy import integrate
def f(x,y):
    return x*y
    pass
def bounds_y():
    return [0,0.5]
    pass
def bound_x(y):
    return [0,1-2*y]
    pass
integrate.nquad(f,[bound_x,bounds_y])
```

## 用sympy 中的diff 求导


```python
import sympy 
sympy.__version__
```

```python
from sympy import diff
from sympy import symbols
def fuc(x):
    return x**4
x=symbols('x')
print(diff(fuc(x),x)) # 必须定义函数 不能使用临时的函数
```
