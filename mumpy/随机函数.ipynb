## 随机函数

numpy.random.rand( )
生成[0,1)之间的数字

```python
import numpy as np
```

```python
np.random.rand(4,2)
```

```python
np.random.rand(2,3,4)
```

```python
print('1dim:',np.random.randint(2,size=5))
print('2-5 dim:',np.random.randint(2,5))
print('2dim:',np.random.randint(-5,5,size=(2,2,2))) #2行2列的数组
```

```python
a=np.random.normal(0,1,(3,2))
b=np.random.normal(0,1,(24,))
print('a',a,'b',b)
```

```python
#和reshape 不一样
a=np.array([[1,2,3],[2,3,4]])
print('a',a)
print('shape',a.shape)
```

```python
b=np.resize(a,(3,3))#第一种用法
b
```

```python
a.resize((3,3),refcheck=False)# 用0填充多的部分
a
```

```python
a=np.array([[1,2,3],[4,5,6]])
print(np.append(a,[7,8]))
print('a',a)
```

```python
print(np.append(a,[[7,8,9]],axis=0))# 0 就是行操作
```

```python
print(np.append(a,[[5],[4]],axis=1))
print(np.append(a,[[2,3,4],[6,6,6]],axis=1))
```

```python
a=np.array([[1,2],[3,4],[5,6]])
print(a)
print(np.insert(a,1,[11,22]))# 在下标为1的位置 插入数据
print(np.insert(a,1,11,axis=0))#在行中插入重复的数字11
print(np.insert(a,1,11,axis=1))#在列中插入重复的数字11
print(np.insert(a,1,[11,12,13],axis=1))
```

```python
a=np.arange(16).reshape(4,4)
print(a)
print('the one:',np.delete(a,14))#删除指定元素 并且返回1维数组
print('row:',np.delete(a,1,axis=0))
print('colomn:',np.delete(a,1,axis=1))
```

### numpy.argwhere()
返回符合条件的 坐标

```python
x=np.arange(12).reshape(3,4)
print(x)
print(x[x>2])
```

```python
print(x)
y=np.argwhere(x>2)
print(y,y.shape)
```

```python
a=np.array([[1,2,3,4],[2,3,4,5],[3,4,5,6]])
print(a)
b=np.unique(a)
print(b)
c,time=np.unique(a,return_counts=True)#   学习一下这个操作
print(c)
print(time)
```

```python
aa=np.array([[1,2,2],[2,2,3]])
cc,times=np.unique(aa,return_counts=True)
print(times)
print('max',np.max(times))
print('argmax',np.argmax(aa))# 返回的是 序号
```
