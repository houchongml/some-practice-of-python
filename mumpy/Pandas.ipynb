# Pandas

```python
import numpy as np
import pandas as pd
```

```python
lis=[12,14,24,11,23]
print(type(lis))
# 使用列表 创建Series
s1=pd.Series(lis)
print(s1)
print(type(s1))
```

```python
s1[1]
```

```python
s1[1]=20#改变 索引为1 的数据值
print(s1)
```

```python
s1[-1] #  出错
```

```python
s1[-1]=-1
print(s1)
```

### 字典 作为数据源 创建Series

```python
d={'a':1,'b':2,'c':3}
se=pd.Series(d)
se
print('index:',se.index)# 标签索引
print('values:',se.values)# 求值
```

```python
se[1]# 可以 用原本的索引 也可以用数字索引
#------条件是 原本的索引有数字的话 就不能用 0，1，2，3。。。索引了！！！
```
