# python 数据结构和算法

```python
#hanoi 问题 用递归算法
def hanoi(n,a,b,c):
    '''
    从a 经过b移动到c
    '''
    if n >0:
        hanoi(n-1,a,c,b)
        print('moving from %s to %s'%(a,c))
        hanoi(n-1,b,a,c)
        pass
```

```python
hanoi(3,'A','B','C')
```

```python
#顺序查找
def linear_search(li,val):
    for ind,v in enumerate(li):
        if v==val:
            return ind
        else:
            return None
        pass
    
```

```python
#二分查找
def binary_search(li,val):
    left=0
    right=len(li)-1
    while left<=right:
        mid=(left+right)//2 
        if li[mid]==val:
            return mid
            pass
        elif li[mid]>val:#待查找的值在mid左侧
            right=mid-1
            pass
        else:#待查找的值在mid右侧
            left=mid+1
            pass
    else:
        return None
    pass

```

```python
li=[1,2,3,4,5,6,7,8,9]
binary_search(li,3)
```

#### 冒泡排序

```python
#冒泡排序
def bubble_sort(li):
    for i in range(len(li)-1):
        for j in range(len(li)-i-1):
            if li[j]>li[j+1]:
                li[j],li[j+1]=li[j+1],li[j]
                pass
            else:
                return
        print(li)

                
```

```python
li=[9,8,7,1,2,3,4,5,6]
print(li)
bubble_sort(li)
print(li)
```

```python
import random
li=[random.randint(0,10000) for i in range(1000)]
print(li)
bubble_sort(li)
print(li)
```

#####  选择排序

```python
#选择排序
def select_sort_simple(li):
    '''
    不推荐，因为复杂度比较高
    '''
    li1=[]
    for i in range(len(li)):
        min_val= min(li)
        li1.append(min_val)
        li.remove(min_val)
        pass
    return li1
```

```python
li=[3,2,4,1,5,6,8,7,9]
print(select_sort_simple(li))
```
