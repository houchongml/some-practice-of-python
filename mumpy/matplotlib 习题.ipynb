#### 在一张图中绘制3组不同的直方图，设置透明度

```python
x1=np.random.normal(0,0.8,1000)# 满足正态分布的数组， 平均值，方差，个数
x2=np.random.normal(-2,1,1000)
x3=np.random.normal(3,2,1000)

plt.hist(x1,alpha=0.3,bins=40,density=True)
plt.hist(x2,alpha=0.3,bins=40,density=False)
plt.hist(x3,alpha=0.3,bins=40,density=False)
```

### 二维直方图

```python
mean=[0,0]
cov=[[1,1],[1,2]]
x,y=np.random.multivariate_normal(mean,cov,10000).T #二维 正态分布数组 
plt.hist2d(x,y,bins=30)
```

### 绘制一张设置网格大小为30 的六角形直方图

```python
plt.hexbin(x,y,gridsize=40)
```

### 绘制平行于x轴的水平参考线

```python
x=np.linspace(0.05,10,100)
y=np.sin(x)
plt.plot(x,y)
plt.axhline(y=0.8,ls='--',c='r')
plt.text(0,0.8,'y=0.8')
```

###  用箭头标出第一个峰值

```python
y=np.linspace(0.05,10,100)
y=np.sin(x)
plt.plot(x,y,'-')
plt.annotate('max',xy=(np.pi/2,1),xytext=(np.pi/2+1,1),weight='bold',color='r',
arrowprops=dict(arrowstyle='->',color='r'))
# xytext 是max的位置
```

```python
fig,ax=plt.subplots()
style=['-','--','-.',':']
x=np.linspace(0,10,1000)
for i in range(4):
    ax.plot(x,np.sin(x+i*np.pi/2),style[i],color='black')
    pass
ax.axis('equal')
```
