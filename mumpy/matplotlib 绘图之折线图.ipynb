2.plt.gca()对坐标轴的操作

get current axes 

```python
x=np.arange(-50,50)
y=x**2
# 获取当前坐标轴
ax=plt.gca()
# 通过坐标轴 spines 确定top  bottom left right 上下左右
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')#去掉上面和右边的框
plt.plot(x,y)
```

data 是移动到具体的刻度

axes 是移动到整个轴的 相对位移 0-1 0 是左边 1是右边

```python
x=np.arange(-50,50)
y=x**2
# 获取当前坐标轴
ax=plt.gca()
# 通过坐标轴 spines 确定top  bottom left right 上下左右
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')#去掉上面和右边的框
#  position 有三种位置参数 data outward axes
#axes 0.0-1.0之间的值 整个轴上面的比例
#ax.spines['left'].set_position(('axes',0.5))
ax.spines['left'].set_position(('data',20))
#移动下轴到指定位置 data 是指按数值挪动 后面的数字 表示挪动到Y轴到刻度值
ax.spines['bottom'].set_position(('data',1000))
plt.plot(x,y)
```

## add_axes():添加区域

数字都是指 画布的相对位置  范围0-1



```python
fig=plt.figure(figsize=(4,2),facecolor='g')#先创建画布
ax1=fig.add_axes([0,0,1,1])#从画布起始位置绘制，宽高和画布一致
ax2=fig.add_axes([0.2,0.2,0.5,0.5])
ax1.plot([1,2,3,4],[2,3,4,5])
ax2.plot([1,2,3,4],[1,4,9,16])
ax3=fig.add_axes([0,0,0.5,0.5])
plt.plot([1,2,3,4,5],[14,15,16,13,15])#用plt 在最近的区域 画
```

### 设置多图的基本信息方式

可以直接在subplot里面设置 各种参数

```python
plt.subplot(211,title='picl',xlabel='x axis')
plt.plot(range(50,70))
plt.subplot(212,title='pic2',xlabel='x axis')
plt.plot(np.arange(10)**2)
```

```python
# 子图标题有重合部分的时候 
plt.subplot(211,title='picl',xlabel='x axis')
plt.plot(range(50,70))
plt.subplot(212,title='pic2',xlabel='x axis')
plt.plot(np.arange(10)**2)
# 关键函数 可以让布局更加紧凑 
plt.tight_layout()
```

```python
fig,axes=plt.subplots(1,2)
#设置 画布的高宽
fig.set_figheight(3)# 实际高度为73*3 像素
fig.set_figwidth(8)
#设置背景
fig.set_facecolor('gray')
# 分别定义x y
x=np.arange(-50,51)
y=x**2
axes[0].plot(x,y)
axes[1].spines['right'].set_color('none')
axes[1].spines['top'].set_color('none')
axes[1].spines['left'].set_position(('axes',0.5))
axes[1].spines['bottom'].set_position(('data',0))
axes[1].plot(x,y)
```
