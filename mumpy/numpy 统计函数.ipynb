## 求平均值 mean()

```python
m=np.arange(20).reshape(4,5)
print(m)
m.mean()
```



0 是每一列的平均    1 是每一行

```python
a=np.array([
    [1,10,3,4],
    [2,3,4,5],
    [3,4,5,6],
    [4,5,6,7]
])
print(a)
a.mean(axis=0)
```

```python
a.mean(axis=1)# 行
```

中位数 np.median

```python
a=([1,2,3,4,5])
np.median(a)
```

标准差 std

```python
a=np.array([2,2,23,3])
print(np.std(a))
print(a.std())
```

var 方差  最大值 max  最小值 min  求和sum

```python
a=np.array([0,2,3,4,5,6]).reshape(2,3)
print(a.var())
print(a.max())
print(a.min())
print(a.sum())
```

max  min 的其他操作 axis

```python
a=np.array([[1,2,3,4],[4,5,6,7],[10,11,12,14]])
print(a)
print('1:{}\n0:{}'.format(a.max(axis=1),a.max(axis=0)))
```

mean 也可以用axis求 行和 以及列和

```python
print(a.sum(axis=0))
```

```python
a=np.array([1,2,3,4],dtype=np.float32)# 浮点型
b=np.array([0,2,3,4],dtype=np.bool_)# 布尔型
print('float{},bool{}'.format(a,b))
```

```python
# str_ 和string 区别
a=np.array([1,2,3,4,5,6])
b=np.float64(a)
c=np.str_(a)
d=np.string_(a)
print(b,c,d)
```

```python
# 创建结构化数据类型
dy=np.dtype([('age','i2')])
print(dy)
#将数据应用于ndarray对象
students=np.array([(18),(19)],dtype=dy)
print(students,students.dtype)
```

```python
teacher=np.dtype([('name',np.str_,2),('age','i1'),('salary','f4')])
## 'name' 名称；  np.str_ 类型；  2 是长度；
print(teacher)
b=np.array([('wl',32.1,8549.1),('hou',20,12323.12),('lin',22,7896.98)],dtype=teacher)
print(b)
```

```python
a=np.array([1,2,3,4],dtype=np.int8)
b=np.array([1,2,3,4],dtype='i8')
## 两种是等价的
print('a:',a.dtype,'b:',b.dtype)
```

```python
b=np.array([127,128,129,-129],dtype=np.int8)
#  int8   -128~~127
b
```

### np.diag()的用法 可以取数组中的对角元素 可以把一维数组 写成矩阵

```python
arr=np.arange(16).reshape([4,4])
print('arr:',arr)
print('diag:',np.diag(arr))#提取元素
brr=np.array([1,2,3,4])
print('diag:',np.diag(brr))#创立矩阵
```
