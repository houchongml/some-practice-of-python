# 重要的区别

```python
a=np.array([1,2,3,4])
b=np.array(a)
#b改变时 a 不变
print('a:',id(a),'b:',id(b))
```

```python
b[1]=10
print('a={},b={}'.format(a,b))
```

```python
a=np.array([1,2,3,4])
b=a
#b改变时 a 变
print('a:',id(a),'b:',id(b))
```

```python
b[1]=10
print('a={},b={}'.format(a,b))
```

```python
np.mat([1,2,3,4])
```

```python
np.mat([[1],[2],[3],[4]])
```

```python
a=np.mat([1,2,3,4])
print(type(a))
at=np.array(a,subok=True)
af=np.array(a)
print('at ID:',id(at),'at type:',type(at),'af ID:',id(af))
```

### resize 调整纬度 如果新数组大于原数组，则填充重复   a.resize 用零填充

```python
a=np.array([[1,2],[2,3]])
print(np.resize(a,(2,3)))
```

```python
a.resize((5,5),refcheck=False)
print(a)
```

### astype 改变数据类型，但是原来的数据不变 

```python
a=np.array([1.1,2.2])
print('a type is :',a.dtype)
print('change type :',a.astype('float32').dtype)
print('a type later :',a.dtype)
```
